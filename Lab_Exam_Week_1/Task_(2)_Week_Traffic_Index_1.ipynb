## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd

df = pd.read_csv('/content/Task (2) Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe().transpose()
```

```python
df.sample(10)
```

```python
df.isnull().sum()
```

```python
df['TravelTimeHistoric']= df['TravelTimeHistoric'].fillna(df['TravelTimeHistoric'].mean())
df['TravelTimeLive']= df['TravelTimeLive'].fillna(df['TravelTimeLive'].mean())
```

```python
df.isnull().sum()
```

### Identify and analyze patterns in the data.

```python
numerical_columns = ['TrafficIndexLive', 'JamsCount', 'JamsDelay', 'JamsLength', 'TrafficIndexWeekAgo', 'TravelTimeHistoric', 'TravelTimeLive']

correlation_matrix = df[numerical_columns].corr()
correlation_matrix
```

```python
df.duplicated().sum()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['Datetime']
```

```python
df['Year'] = df['Datetime'].dt.year
df['Year']
```

#### This Bar Plot Represent The Total Delay in Different Cities

**We can take insights from it by knowing what is the city that has a bigger delay time than others**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='JamsDelay', data=df)
plt.xlabel('City')
plt.ylabel('Total Delay in Minutes')
plt.title('Total Delay in Different Cities')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


**This Heatmap Represent The *Variables* That Has Realationships**

we can see the varibles (Jam_dellay) and (Jam_count) have power relationship

```python
sns.heatmap(correlation_matrix, annot=True)
plt.title('The Relationship Between Variables')
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.
